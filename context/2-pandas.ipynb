## Combinando DataFrames

### Merge y Concat

```python
df1 = pd.DataFrame(
    {'A':['A0','A1','A2','A3'],
    'B':['B0','B1','B2','B3'],
    'C':['C0','C1','C2','C3'],
    'D':['D0','D1','D2','D3']
    }
)
df1
```

```python
df2 = pd.DataFrame(
    {'A':['A4','A5','A6','A7'],
    'B':['B4','B5','B6','B7'],
    'C':['C4','C5','C6','C7'],
    'D':['D4','D5','D6','D7']
    }
)
df2
```

```python
pd.concat([df1,df2], ignore_index=True)
```

```python
pd.concat([df1,df2], axis=1)
```

```python
izq = pd.DataFrame({'key':['k0','k1','k2','k3'],
 'A' : ['A0', 'A1','A2','A3'],
 'B' : ['B0','B1','B2','B3']
})

der = pd.DataFrame({'key':['k0','k1','k2','k3'],
 'C' : ['C0', 'C1','C2','C3'],
 'D' : ['D0','D1','D2','D3']
})
```

```python
izq
```

```python
der
```

```python
izq.merge(der, on='key')
```

```python
izq = pd.DataFrame({'key':['k0','k1','k2','k3'],
 'A' : ['A0', 'A1','A2','A3'],
 'B' : ['B0','B1','B2','B3']
})

der = pd.DataFrame({'key1':['k0','k1','k2','k3'],
 'C' : ['C0', 'C1','C2','C3'],
 'D' : ['D0','D1','D2','D3']
})
```

```python
izq.merge(der, left_on='key', right_on='key1')
```

```python
izq = pd.DataFrame({'key':['k0','k1','k2','k3'],
 'A' : ['A0', 'A1','A2','A3'],
 'B' : ['B0','B1','B2','B3']
})

der = pd.DataFrame({'key1':['k0','k1','k2',np.nan],
 'C' : ['C0', 'C1','C2','C3'],
 'D' : ['D0','D1','D2','D3']
})
```

```python
izq.merge(der, left_on='key', right_on='key1')
```

```python
# LEFT JOIN
izq.merge(der, left_on='key', right_on='key1', how='left')
```

```python
# INNER JOIN 
izq.merge(der, left_on='key', right_on='key1', how='inner')
```

```python
# RIGHT JOIN
izq.merge(der, left_on='key', right_on='key1', how='right')
```

### Join

El Join va directamente a los indices con los que estemos trabajando los DF's, no ha columnas especificas

```python
izq = pd.DataFrame({'A': ['A0','A1','A2'],
  'B':['B0','B1','B2']},
  index=['k0','k1','k2'])

der =pd.DataFrame({'C': ['C0','C1','C2'],
  'D':['D0','D1','D2']},
  index=['k0','k2','k3']) 
```

```python
izq
```

```python
der
```

```python
# LEFT JOIN
izq.join(der, how='left')
```

```python
# RIGHT JOIN
izq.join(der, how='right')
```

```python
# INNER JOIN
izq.join(der, how='inner')
```

```python
# OUTER JOIN
izq.join(der, how='outer')
```
